```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.datasets import load_boston

```

# Question #1) Import Housing Data

```python
boston = load_boston()
bos = pd.DataFrame(boston.data)
print(bos.head())
```

```python
print(boston.DESCR)
```

```python
bos.columns = boston.feature_names
print(bos.head())
```

```python
bos['PRICE'] = boston.target
print(bos.head())
```

# Question #2) Scatter Plots

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
for i,feature in enumerate(['CRIM', 'ZN', 'INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']):
    plt.figure(i)
    sns.scatterplot(x=bos[feature], y=bos['PRICE']);
#plt.xlabel("Prices: $Y_i$")
#plt.ylabel("Predicted prices: $\hat{Y}_i$")
```

# Question #3) Linear Regression w/ R2 scores


```python

from sklearn import preprocessing
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = bos.drop('PRICE', axis = 1)
Y = bos['PRICE']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state=30)
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score
y_train_predict = lin_model.predict(X_train)
r2 = r2_score(Y_train, y_train_predict)
print('R2 linear regression training data score is {}'.format(r2))
```

```python
y_test_predict = lin_model.predict(X_test)
r2 = r2_score(Y_test, y_test_predict)
```

```python
print('R2 linear regression testing data score is {}'.format(r2))
```

# Question #4) Polynomial Regession w/ R2 scores

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_poly,Y, test_size = 0.20, random_state = 5)
```

```python
from sklearn.preprocessing import StandardScaler
#standard scaler (fit transform on train, fit only on test)
s_scaler = StandardScaler()
X_train = s_scaler.fit_transform(X_train.astype(np.float))
X_test= s_scaler.transform(X_test.astype(np.float))

#fit model
poly_lr = LinearRegression().fit(X_train, Y_train)
y_train_predict = poly_lr.predict(X_train)
r2 = r2_score(Y_train, y_train_predict)
print('R2 polynomial regression training data score is {}'.format(r2))

```

```python
y_test_predict = poly_lr.predict(X_test)
r2 = r2_score(Y_test, y_test_predict)
print('R2 polynomial regression test data score is {}'.format(r2))
```
